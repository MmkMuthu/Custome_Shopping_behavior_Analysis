```python
import pandas as pd
```

```python
data = pd.read_csv("customer_shopping_behavior_raw.csv")
```

```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
data = data.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
data.columns
```

```python
data['review_rating'] = data.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
labels = ['Young','Adult','Middle-aged','Senior']
data['age_group'] = pd.qcut(data['age'],q=4,labels = labels)
```

```python
data[['age_group','age']].head()
```

```python
(data['discount_applied'] == data['promo_code_used']).all()
```

```python
data = data.drop('promo_code_used', axis=1)
```

```python
data.columns
```

```python
data['frequency_of_purchases'].unique()
```

```python
Mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually' : 365,
    'Quarterly' : 91,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Every 3 Months' : 90,
}

data['frequency_of_days'] = data['frequency_of_purchases'].map(Mapping)
```

```python
data[['frequency_of_days','frequency_of_purchases']].head(10)
```

```python
data.to_csv("customer_shopping_behavior_cleaned.csv",index=False)
```
